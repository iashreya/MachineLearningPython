```python
#Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
yoe = dataset.iloc[:, :-1].values
sal = dataset.iloc[:, 1].values
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(yoe, sal)
```

```python
y_pred = regressor.predict(yoe)
```

```python
plt.plot(yoe, sal, color = 'red')
plt.plot(yoe, y_pred, color = 'blue')
plt.show()
```

```python
import numpy as np
```

```python
dataset = np.genfromtxt('Salary_Data.csv', delimiter=',')
```

```python
dataset = dataset[1:]
yoe = dataset[0:,0]
sal = dataset[0:, 1]
```

```python
#Variables for the model
sum_yoe = np.sum(yoe)
sum_sal = np.sum(sal)
sum_yoe_sq = np.sum(yoe*yoe)
sum_yoe_sal = np.sum(yoe*sal)

change = 1
theta0 = 0
theta1 = 0
m = len(dataset)
alpha = 0.008

while (change == 1):
  temp0 = theta0 - (alpha/m)*(m*theta0 + theta1*sum_yoe - sum_sal)
  temp1 = theta1 - (alpha/m)*(theta0*sum_yoe + theta1*sum_yoe_sq - sum_yoe_sal)
  if(temp0 == theta0 and temp1 == theta1):
    change = 0
  theta0 = temp0
  theta1 = temp1
  theta0
  theta1
```

```python
def fun(x): return(theta0 + theta1*x)
```

```python
plt.plot(yoe, sal, color = 'red')
plt.plot(yoe, fun(yoe), color = 'green')
plt.show()
```

```python
#Correlation between my predictions and the in-built regression model
np.corrcoef(y_pred, fun(yoe))
```
